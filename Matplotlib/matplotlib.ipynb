## Data Visualization With MATPLOTLIB

`Matplotlib` - An object-oriented plotting python library. Matplotlib is the Data Visualization tool in Python, in which you can plot amazing charts. The major feature of Matplotlib is ability to plot graphics backends, which supports dozen of charts that includes Bar chart, Pie Chart, Histogram, Scatter Plot, Line Charts and many more.

### Figure and Axes

The term figure in Matplotlib is basically a container that contains all the objects that is used to represent graphs(plots),axes, text, title, and label.
And the term axes is basically a grid with ticks and labels, that is used to plot the elements. 

````python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes()
plt.show() 
````

Figsize is an argument passed inside plt.figure.

````python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(15,12))
ax = plt.axes()
plt.show()
````

Let us enable the grid for the axes.

````python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes()
ax.grid("on")
plt.show() 
````

### Line Plot

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,15,150)  #we have discussed linspace in Day1: Numpy go check out that to understand this command
plt.plot(np.sin(x),label="Sine Wave",color="g")
plt.plot(np.cos(x),label="Cosine Wave",color="orange")
plt.legend(loc="upper right")
plt.show()
````

Line Plot is used analyze how the sales of the companies develops in a year, or stock market growth graph. 
Basically Line plot is preferred when the data changes over Time. 

#### What Is Legend?

A legend is an area that represent the category of the graph.With help of legend method it becomes easier to create labels for the plots which eventually helps us differentiating the functions drawn in the plot. 

#### Color And Marker
To decorate your graph

Commonly Used Color:
- k=black
- r=red
- b=blue
- g=green
- y=yellow
- orange

Commonly Used Marker:
- '+'
- '.'
- '_'
- '*'
- '^'
- 'D' - Diamond
- 'H' - hexagon

````python
x = np.linspace(0,15,150)
plt.plot(np.sin(x),label="Sine Wave",color="r",marker="H")
plt.plot(np.cos(x),label="Cosine Wave",color="b",marker="D")
plt.legend(loc="upper right")
plt.show()
````

### Labels and Title

````python
plt.plot(np.sin(x),label="Sine Wave",color="violet",marker="^")
plt.plot(np.cos(x),label="Cosine Wave",color="aqua",marker="+")
plt.legend(loc="upper right")
plt.title("Trignometry Function")
plt.xlabel("x")
plt.ylabel("sin(x),cos(x)")
plt.show()
````

### Bar Plot

Bar graphs are the pictorial representation of data (generally grouped), in the form of vertical or horizontal rectangular bars, where the length of bars are proportional to the measure of data. 

````python
import matplotlib.pyplot as plt
import numpy as np

anime_names = ["One Piece","Naruto","Bleach","Gintama","Fairy Tail"]
episodes = [1000,720,366,360,180]
plt.bar(anime_names,episodes)
plt.xlabel("Anime Names")
plt.ylabel("Total Episodes")
plt.show()
````

### Pie Plot

````python
marks = np.array([92,91,90,85,81])
subjects = ["Computer","Maths","Electronics","English","Management"]
plt.pie(marks,labels=subjects,shadow=True,explode=[0,0,0.2,0,0.15],autopct="%0.3f%%")
plt.show()
````

1. shadow

By default Shadow is False. By enabling Shadow to True, a greyish surface below the pie chart appears that makes pie chart look attractive. 

2. autopct

Suppose we have divided pie chart into various categories. By assigning autopct to certain float value we can add a label on individual categories to check its overall spread.

We can manually calculate autopct, it is no big deal.

plt.pie([92,95,80,90,75],labels=["Maths","Computer","Electronics","English","Science"],autopct="%0.3f%%")

Since Maths is 92, and overall sum of our data is 432.
The overall spread of Mathematics is (92*100)/432. In the same way to rest of the data. When we check the overall representation of Maths we get 21.296296296296298, since autopct = 0.3f, it will stop at 3rd decimal point and be labelled inside the pie chart. 

3. explode

explode is a list that is equal to the length 

4. labels

Providing labels to Pie chart is must. If label argument is missed then understanding pie chart is impossible. Labels give meaning to pie plot. And labels is represented by a list whose length should match that with the data. 

### Scatter Plot

Scatter Plot is used to compare between two features, i.e., to check relationship between two or more variable. 

````python
x = [12,45,23,10,30]
y = [20,17,34,3,10]
plt.scatter(x,y,color="red")
plt.show()
````

````python
#You can even add more Scatter Plot in one graph and provide legend for better understanding. 

x = [12,45,23,10,30]
y = [20,17,34,3,10]
plt.scatter(x,y,color="red",marker="+",label="X to Y")
plt.scatter(y,x,color="blue",marker="*",label="Y to X")
plt.legend()
````

### Subplot

Subplot is used to plot multiple graphs in one single figure windows by sub-dividing into sub categories. 


````python
anime_names = ["One Piece","Naruto","Bleach","Gintama","Fairy Tail"]
episodes = [1000,720,366,360,180]

plt.subplot(2,1,1)
plt.bar(anime_names,episodes)
plt.ylabel("Total Episodes")

plt.subplot(2,1,2)
plt.bar(anime_names,episodes,color=["red","yellow","black","green","blue"])
plt.xlabel("Anime Names")
plt.ylabel("Total Episodes")
plt.show()
````

### How to save the plots locally?

You are doing good in plotting different graphs. But, Is that enough?You need to know how to save this files into your local system. Matplotlib also enables us to save the chart/graphs which we have plotted.

````python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,15,150)
plt.plot(np.sin(x),label="Sine Wave",color="g")
plt.plot(np.cos(x),label="Cosine Wave",color="k")
plt.legend(loc="upper right")
plt.savefig('trigono.png')
plt.show()
````
